```python
first_name = []
last_name = []
```

```python
for i in range(1000):
    first_name.append(names.get_first_name())
    last_name.append(names.get_first_name())
```

```python
df.user_name, df.usersurname = first_name, last_name
```

```python
sufijos = ['@gmail.com', '@hotmail.com', '@outlook.com', "@aol.com", "@mail.com" , "@mail.kz", '@yahoo.com']
emails = []
for i in range(1000):
       #char_num = np.random.randint(1,10)
       email = df.user_name[i] + '.' + df.usersurname[i] + str(np.random.randint(1,10)) + random.choice(sufijos)
       emails.append(email)
```

```python
df.email = emails
```
