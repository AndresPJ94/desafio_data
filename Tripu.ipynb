```python
import random
import string

def random_char(char_num):
       return ''.join(random.choice(string.ascii_letters) for _ in range(char_num))

print (random_char(10)+"@gmail.com")
```

```python
def idioma_multiple():
    prob = np.random.randint(2)
    
    if prob == 0:
        idi1= np.random.choice(idioma)
        return idi1
    else:
        idi2 = np.random.choice(idioma,2, replace=False)
        return  idi2
idioma_multiple()
```

```python
database=pd.DataFrame(columns=['ID', 'Nombre', 'Apellidos', 'Email', 'Password',
       'Año de Nacimiento', 'Genero', 'Idioma', 'Años en España',
       'Nivel de Estudios', 'Trabajo', 'Tipo de Apoyo', "Experto",
       'Comarcas (Sierra Norte)', "Sobre mi", 'Fecha de Registro'])


users=1000

for a in range(users):

    id=a+1
    mail=random_char(10)+"@gmail.com"
    name=random_char(7)
    surname=random_char(8)
    password=random_char(9)
    born=random.randrange(1964,2004)
    gen=np.random.choice(genero)
    idi=idioma_multiple() # multichoice maximo 2 
    years=np.random.choice(yearsInSpain)
    studies=np.random.choice(estudios)
    work=np.random.choice(trabajo)
    ayuda=np.random.choice(apoyo)
    exp=np.random.choice(experto)
    # sobre=np.random.choice(bout) 
    sobre = random.sample(list(bout.value_counts().keys()), 3) # multichoice maximo 3 
    lug=np.random.choice(comarcas)
    
    start_date = datetime.date(2021, 1, 1)
    end_date   = datetime.date(2023, 1, 12)
    num_days   = (end_date - start_date).days
    rand_days   = random.randint(1, num_days)
    random_date = start_date + datetime.timedelta(days=rand_days)
    fecha_reg=random_date



    database.loc[a]=[id, name, surname, mail, password, born, gen, idi, years, studies, work, ayuda, exp, lug, sobre, fecha_reg]


database.head()
```

```python
database_copia = database.copy()
```

```python
def list_to_string(columna):
    for i,a in enumerate(columna):
        if type(a) != str:
            texto = ', '.join(columna[i])
            columna[i] = texto
```

```python
list_to_string(database_copia['Idioma'])
```
